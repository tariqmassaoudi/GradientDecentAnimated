## EXERCICE 3 : OPTIMISATION D'UNE FONCTION A TROIS VARIABLES. APPLICATION AU RECALAGE D'IMAGES RIGIDE


Le but de cet exercice est d'utiliser une fonction d'estimation des gradients d'une energie de recalage rigide pour coder une descente de gradient.
    
L'algorithme sera utilisé sur les images Brains_Fixed.png et Brains_Moving.png

Le point le plus interessant à décrire et à justifier sera comment l'algorithme est parametré.



-------------------------------------------------------------------------------------

##### EXEMPLE D'UTILISATION DES FONCTIONS DE RegLib ET FAMILIARISATION AVEC LE RECALAGE RIGIDE

On va deformer l'image ImFix pour qu'elle ressemble le plus possible à l'image ImMov.

Pour cela :
- On utilise RegLib.TranslateAndRotateImage(RCX,RCY,theta,dx,dy,ImMov,ImDef) qui va transformer l'image ImMov en ImDef avec une rotation d'angle theta autour du point (RCX,RCY) puis une translation de (dx,dy) pixels
- On mesure à quel point ImDef et  ImFix se ressemblent avec RegLib.Cpt_SSD(ImDef,ImFix). Cette mesure de SSD est la somme des differences d'intensite au carre entre les points correspondant (elle est mathématiquement equivalente à la Sum of Squared Error en apprentissage machine).


```python
import numpy as np
import RegLib

ImFix=RegLib.OpImage('Brains_Fixed.png')
ImDef=RegLib.OpImage('Brains_Fixed.png')  #... pour s'assurer que l'image Deformee soit dans le meme domaine que l'image fixe
ImMov=RegLib.OpImage('Brains_Moving.png')

ImMov.show()


RegLib.TranslateAndRotateImage(10,10,0.0,0.,0.,ImMov,ImDef)
print('sqrt(SSD)='+str(np.sqrt(RegLib.Cpt_SSD(ImDef,ImFix))))
ImDef.show()

RegLib.TranslateAndRotateImage(10,10,-0.1,-20.,30.,ImMov,ImDef)
print('sqrt(SSD)='+str(np.sqrt(RegLib.Cpt_SSD(ImDef,ImFix))))
ImDef.show()


ImFix.SaveComparisonWithAnotherImage(ImDef,LabelSelfIm='Fixed image',LabelComparedIm='Deformed image',filename='ComparedImages.png')
```


#### EXERCICE : 
Pour recaler au mieux les images, nous allons trouver la rotation et la translation pour laquelle la SSD est minimisée. Nous allons alors effectuer une descente de gradient dans laquelle la fonction de coût est SSD et les 3 parametres sont (theta,dx,dy).


```python


import numpy as np
import RegLib

#1.1) lit les images initiales
ImFix=RegLib.OpImage('Brains_Fixed.png')
ImDef=RegLib.OpImage('Brains_Fixed.png')  #... pour s'assurer que l'image Deformee soit dans le meme domaine que l'image fixe
ImMov=RegLib.OpImage('Brains_Moving.png')

```

```python
#1.2) fonction d'estimation du gradient d'energie
def CptGradE(ImFix,ImDef,ImMov,Cx,Cy,theta,Delta_theta,dx,dy,Delta_dxdy):
  """
  Approximate the gradient of E with respect to theta, dx and dy using centered finite differences.
  Inputs:
    ImFix,ImDef,ImMov: The fixed, moving and deformed images
    Cx,Cy: rotation center
    theta: rotation in radians 
    Delta_theta: Delta rotation for the finite differences
    dx,dy: translation on the x and y axes
    Delta_dxdy: Delta translation for the finite differences
  Outputs:
    dEdtheta,dEddx,dEddy: Gradient of E with respect to theta, dx and dy
  """
  #1.2.1) estimate the energy derivative with respect to theta
  RegLib.TranslateAndRotateImage(Cx,Cy,theta+Delta_theta,dx,dy,ImMov,ImDef)
  E_thetaPlus=RegLib.Cpt_SSD(ImDef,ImFix)
  RegLib.TranslateAndRotateImage(Cx,Cy,theta-Delta_theta,dx,dy,ImMov,ImDef)
  E_thetaMinus=RegLib.Cpt_SSD(ImDef,ImFix)
  dEdtheta=(E_thetaPlus-E_thetaMinus)/(2*Delta_theta)
  
  #1.2.2) estimate the energy derivative with respect to dx
  RegLib.TranslateAndRotateImage(Cx,Cy,theta,dx+Delta_dxdy,dy,ImMov,ImDef)
  E_dxPlus=RegLib.Cpt_SSD(ImDef,ImFix)
  RegLib.TranslateAndRotateImage(Cx,Cy,theta,dx-Delta_dxdy,dy,ImMov,ImDef)
  E_dxMinus=RegLib.Cpt_SSD(ImDef,ImFix)
  dEddx=(E_dxPlus-E_dxMinus)/2.
  
  #1.2.3) estimate the energy derivative with respect to dy
  RegLib.TranslateAndRotateImage(Cx,Cy,theta,dx,dy+Delta_dxdy,ImMov,ImDef)
  E_dyPlus=RegLib.Cpt_SSD(ImDef,ImFix)
  RegLib.TranslateAndRotateImage(Cx,Cy,theta,dx,dy-Delta_dxdy,ImMov,ImDef)
  E_dyMinus=RegLib.Cpt_SSD(ImDef,ImFix)
  dEddy=(E_dyPlus-E_dyMinus)/2.
  
  return dEdtheta,dEddx,dEddy
```

```python


#1.3) parametres initiaux
theta=0.
dx=0.
dy=0.

CX=180   #Centre de rotation sur l'axe x pre-defini
CY=180   #Centre de rotation sur l'axe y pre-defini  
RegLib.TranslateAndRotateImage(180,180,theta,dx,dy,ImMov,ImDef)
E_init=RegLib.Cpt_SSD(ImDef,ImFix)
print('Initial energy:'+str(E_init))


ImFix.CompareWithAnotherImage(ImMov,LabelSelfIm='Fixed image',LabelComparedIm='Moving image',ShowAll=1)
```

```python

#1.4) descente de gradient -> a faire et justifier

[dEdtheta,dEddx,dEddy]=CptGradE(ImFix,ImDef,ImMov,180,180,theta,0.017,dx,dy,1.)

epsilon_Theta=np.fabs(0.017/dEdtheta)*0.5         #pour controler la vitesse de convergence en theta
epsilon_dx=np.fabs(1./dEddx)*0.5                  #pour controler la vitesse de convergence en dx
epsilon_dy=np.fabs(1./dEddy)*0.5                  #pour controler la vitesse de convergence en dy


for i in range(30):
  print('Iteration '+str(i))
  #4.1) calcul du gradient d'energie   (pas de regularisation des deformations ici car on ne fait pas d'hypothese sur ce que devrait etre la deformation ideal)
  [dEdtheta,dEddx,dEddy]=CptGradE(ImFix,ImDef,ImMov,180,180,theta,0.017,dx,dy,1.)
  
  #4.2) mise a jour des parametres
  theta=theta-dEdtheta*epsilon_Theta
  dx=dx-dEddx*epsilon_dx
  dy=dy-dEddy*epsilon_dy
  
  #4.3) affiche et sauvegarde l'evolution des parametres
  E=RegLib.Cpt_SSD(ImDef,ImFix)
  print("theta="+str(theta)+"  dx="+str(dx)+"  dy="+str(dy)+"  -> E="+str(E))
  
  RegLib.TranslateAndRotateImage(180,180,theta,dx,dy,ImMov,ImDef)
  name='DefImage_Iteration'+str(i)+'.png'
  ImFix.SaveComparisonWithAnotherImage(ImDef,LabelSelfIm='Fixed image',LabelComparedIm='Deformed image',filename=name)

```

```python
#2) resultat final
RegLib.TranslateAndRotateImage(180,180,theta,dx,dy,ImMov,ImDef)
print("Final parameters: theta="+str(theta)+",  dx="+str(dx)+",  dy="+str(dy))
print('Final energy:'+str(E))

ImFix.CompareWithAnotherImage(ImDef,LabelSelfIm='Fixed image',LabelComparedIm='Deformed image',ShowAll=1)
```
