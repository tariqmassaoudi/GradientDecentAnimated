## TP optimisation -- Exercice 1 : Descente de gradient en 1D

Nous allons ici programmer à la main des algorithmes qui nous permettent de minimiser des fonctions en dimension 1 ou n. Aucune contrainte ne sera considérée dans ce premier exercice.



Plus spécifiquement, on souhaite minimiser $f(x) = \frac{1}{4} x^4 - (x-0.7)^2$ à l'aide d'une methode de descente de gradient, i.e. :
- **Init :** $x_0 = x_{init}$
- **Repeat until convergence :** $x_{i+1} = x_{i} - \alpha \nabla f (x_i)$

où le gradient de $f$ (i.e. $\nabla f$) correspond ici à sa derivé par rapport à $x$. Elle pourra être calculee soit analytiquement soit par differences finies





```python
#import libraries
import numpy as np
import matplotlib.pyplot as plt
```

#### Questions 1 :
- Calculez sur feuille le minimum de cette fonction (où sa derivée s'annule)
- Codez funct(x) qui retourne f(x)
- Codez Grad_funct(x,epsilon) qui retourne le gradient de f au point x. Le calcul se fera par différences finies, i.e. sera approxime par (funct(x+epsilon)-funct(x))/epsilon ou epsilon est petit
- Utilisez Grad_funct(x) pour minimiser f(x). On prendra typiquement un x_init compris entre -3 et 3. On prendra aussi par exemple alpha=0.1 et epsilon=0.0001
- Représenter la convergence de l'algorithme, i.e. abs(x_{final} - x_{i}) pour $i \in [0,1,...,final]$.



```python
#definition de funct
def funct(x):
  return 0.25*np.power(x,4.)-np.power(x-0.7,2.)

grid_x=np.linspace(-3,3,100)
plt.plot(grid_x,funct(grid_x))
plt.show()
```

```python
#definition de Grad_funct

def Grad_funct(x,epsilon):
  return (funct(x+epsilon)-funct(x))/epsilon


alpha=0.1
x=-3+6*np.random.rand()
list_evo_x=[x]
for i in range(100):
  x=x-alpha*Grad_funct(x,0.0001)
  #print(x)
  list_evo_x.append(x)
  
plt.plot(list_evo_x)
plt.show()
```

#### Questions 2 :
- Fixons x_ini=0.5, epsilon=0.0001 et arretons les caculs à 100 iterations. Visualisez la convergence de l'algorithme quand alpha vaut [0.001,0.01,0.1,1.,10.]
- Comment interpretez-vous les differents resultats



```python

alpha=0.001
x=0.5
list_evo_x=[x]
for i in range(100):
  x=x-alpha*Grad_funct(x,0.0001)
  #print(x)
  list_evo_x.append(x)
plt.plot(list_evo_x)
plt.show()
```

-> trop lent

```python
alpha=0.01
x=0.5
list_evo_x=[x]
for i in range(100):
  x=x-alpha*Grad_funct(x,0.0001)
  #print(x)
  list_evo_x.append(x)
plt.plot(list_evo_x)
plt.show()
```

-> trop lent


```python

alpha=0.1
x=0.5
list_evo_x=[x]
for i in range(100):
  x=x-alpha*Grad_funct(x,0.0001)
  #print(x)
  list_evo_x.append(x)
plt.plot(list_evo_x)
plt.show()
```

-> bon


```python
alpha=1.
x=0.5
list_evo_x=[x]
for i in range(100):
  x=x-alpha*Grad_funct(x,0.0001)
  #print(x)
  list_evo_x.append(x)
plt.plot(list_evo_x)
plt.show()

```

-> diverge

```python
alpha=10.
x=0.5
list_evo_x=[x]
for i in range(100):
  x=x-alpha*Grad_funct(x,0.0001)
  #print(x)
  list_evo_x.append(x)
plt.plot(list_evo_x)
plt.show()
```

-> diverge encore plus


**conclusion :** le choix du alpha est très important et doit tenir compte des donnees. Il faut du bon sens.




### Questions 3 :
- Fixons x_ini=0.5, alpha=0.1 et arretons les caculs a 100 iterations. Visualisez la convergence de l'algorithme quand epsilon vaut [10,1,10e-1,10e-5,10e-10,10e-20]
- Comment interpretez-vous les differents resultats



```python
alpha=0.1
x=0.5
list_evo_x=[x]
for i in range(100):
  x=x-alpha*Grad_funct(x,10)
  #print(x)
  list_evo_x.append(x)
plt.plot(list_evo_x)
plt.show()
```

-> diverge -- mauvaise approximation de la derivée


```python

alpha=0.1
x=0.5
list_evo_x=[x]
for i in range(100):
  x=x-alpha*Grad_funct(x,1)
  #print(x)
  list_evo_x.append(x)
plt.plot(list_evo_x)
plt.show()
```

-> pas très précis mais on y arrive

-> derivées à plat - on attent les limites de precision des floats (ou double)



**Conclusion :** epsilon est plus simple à paramétrer que alpha mais il ne doit pas être trop grand pour bien approximer une derivée et ne pas être trop petit par rapport à la précision d'encodage des valeurs
